# distances

```python
def compare_two_columns_across(df_1, df_2, df_1_col, df_2_col, prefixes):
    collection = []
    for i, df_1_row in df_1.iterrows():
        for j, df_2_row in df_2.iterrows():
            try:
                dist = distance(df_1_row[df_1_col], df_2_row[df_2_col])
            except Exception as e:
                import pdb; pdb.set_trace()
            
            data = {
                f'{prefixes[0]}_col': df_1_row[df_1_col],
                f'{prefixes[1]}_col': df_2_row[df_2_col],
                f'{prefixes[0]}_index': i,
                f'{prefixes[1]}_index': j,
                'distance': dist
            }

            collection.append(data)

    return pd.DataFrame(collection)
```

```python
mask_name_distances = compare_two_columns_across(
    mask_nerd_kids, 
    certainly_high_quality_filters_armbrust, 
    'unique_internal_model_code',
    'unique_internal_model_code',
    prefixes=('mnk', 'armbrust')
)
```

```python
mask_name_distances
```
