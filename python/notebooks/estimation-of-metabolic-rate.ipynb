```python
with_dummies = pd.get_dummies(reset, 'age')
with_dummies = with_dummies.rename(columns={'age_man': 'man'})
```

```python
with_dummies.loc[:, 'met'] = with_dummies['met'] / 10.0
```

```python
with_dummies.columns
```

```python
predictor_column_names = set(with_dummies.columns) - {
    'BMR (MJ/day)', 
    'mean body mass (kg)', 
    'carbon_dioxide_generation_rate',
    'age_woman',
    'age_<1'
}
```

```python
predictor_column_names
```

carbon dioxide generation level = 
    I(age < 1) * I(man) ( A  * met + C ) + 
    I(age < 1) * (1 - I(man)) ( B * met + D ) +
    I(1<age<3) * I(man) ( A  * met + C ) + 
    I(1<age<3) * (1 - I(man)) ( B * met + D ) + ...

```python
with_dummies.columns
```

```python

ages = with_dummies.columns[with_dummies.columns.str.contains('age_') & (~with_dummies.columns.isin(['age_woman']))]
```

```python
ages
```

```python
with_dummies
```

```python
round(1.2)
```

```python
results = {
    0: {}, # woman
    1: {} # man
}

for age in ages:
    for man in [0, 1]:
        
        sub_df = with_dummies[(with_dummies[age] == True) & with_dummies['man'] == man]
        lr = LinearRegression()
        lr.fit(sub_df[['met']], sub_df['carbon_dioxide_generation_rate'])
        
        results[man][age] = {
            'coef': round(lr.coef_[0], 6), 
            'intercept': round(lr.intercept_, 6)
        }
```

```python
results
```

```python
def predict(mets, age_group, man, model):
    dictionary = model[man][age_group]
    return dictionary['coef'] * mets + dictionary['intercept']
```

```python
fig, ax = plt.subplots(1,1, figsize=(10,10))
mets = np.array([1.0, 1.2, 1.4, 1.6, 2.0, 3.0, 4.0])
age = 'age_11 to <16'
man = 1                
y_pred = predict(mets, age, man, results)
                
ax.plot(mets, y_pred, label='pred')
ax.plot(mets, with_dummies[(with_dummies[age] == True) & with_dummies['man'] == man]['carbon_dioxide_generation_rate'], label='true')
ax.legend()
```
